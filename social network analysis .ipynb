```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import warnings; warnings.simplefilter('ignore')
```

```python
G_symmetric = nx.Graph()

G_symmetric.add_edge('sohaib','abdullah',weight=25)
G_symmetric.add_edge('sohaib','saif',weight=6)
G_symmetric.add_edge('sohaib','zain',weight=15)
G_symmetric.add_edge('zain','saif',weight=15)
G_symmetric.add_edge('abdullah','saif',weight=8)
G_symmetric.add_edge('saif','sohaib',weight=6)
G_symmetric.add_edge('zain','sohaib',weight=20)
G_symmetric.add_edge('zain','abdullah',weight=7)
G_symmetric.add_edge('sohaib','haseeb',weight=11)
G_symmetric.add_edge('haseeb','noman',weight=14)
G_symmetric.add_edge('haseeb','kamran',weight=16)
G_symmetric.add_edge('kamran','noman',weight=3)
G_symmetric.add_edge('sohaib','ayesha',weight=6)
G_symmetric.add_edge('ayesha','mahnoor',weight=6)
G_symmetric.add_edge('ayesha','mishal',weight=15)
G_symmetric.add_edge('mahnoor','mishal',weight=16)
G_symmetric.add_edge('haseeb','abdullah',weight=7)
```

```python
print(nx.info(G_symmetric))
```

```python
plt.figure(figsize=(5,5))
nx.draw_networkx(G_symmetric);
```

```python
elarge = [(u, v) for (u, v, d) in G_symmetric.edges(data=True) if d['weight'] > 8]
esmall = [(u, v) for (u, v, d) in G_symmetric.edges(data=True) if d['weight'] <= 8]

pos = nx.circular_layout(G_symmetric)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G_symmetric, pos, node_size=700)

# edges
nx.draw_networkx_edges(G_symmetric, pos, edgelist=elarge,width=6)
nx.draw_networkx_edges(G_symmetric, pos, edgelist=esmall,width=6, alpha=0.5, edge_color='b', style='dashed')

# labels
nx.draw_networkx_labels(G_symmetric, pos, font_size=16, font_family='sans-serif')

plt.axis('off')
plt.show();
```

```python
pos = nx.spring_layout(G_symmetric)
betCent = nx.betweenness_centrality(G_symmetric, normalized=True, endpoints=True)
sorted(betCent, key=betCent.get, reverse=True)[:5]
node_color = [20000.0 * G_symmetric.degree(v) for v in G_symmetric]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G_symmetric, pos=pos, with_labels=True,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off');
```

```python
pos = nx.spring_layout(G_symmetric)
betCent = nx.closeness_centrality(G_symmetric)
sorted(betCent, key=betCent.get, reverse=True)[:5]
node_color = [20000.0 * G_symmetric.degree(v) for v in G_symmetric]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G_symmetric, pos=pos, with_labels=True,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off');
```
